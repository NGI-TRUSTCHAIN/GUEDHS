```python
# third party
import pandas as pd

# syft absolute
import syft as sy
```

```python
chuc_client = sy.login(port=8081, email="info@openmined.org", password="changethis")
chudsa_client = sy.login(port=8082, email="info@openmined.org", password="changethis")
```

```python
virus = sy.DataSubject(
    name="Virus",
    aliases=["virus"],
    description="Submicroscopic infectious agent that replicates only inside the living cells of an organism.",
)

rsv = sy.DataSubject(
    name="RSV",
    aliases=["virus:rsv"],
    description="Respiratory syncytial virus (RSV) is a common, and very contagious, virus that infects the respiratory tract of most children before their second birthday",
)

virus.add_member(rsv)
```

```python
chuc_client.data_subject_registry.add_data_subject(virus)
chudsa_client.data_subject_registry.add_data_subject(virus)
```

```python
assert len(chuc_client.data_subject_registry.get_all()) == 2
assert len(chudsa_client.data_subject_registry.get_all()) == 2
```

```python
chuc_dataset = sy.Dataset(name="RSV Dataset", description="Dataset with RSV data")

chuc_rsv_data = pd.read_csv("chuc.csv")
chuc_rsv_mock = chuc_rsv_data.sample(frac=0.1)

chuc_rsv_asset = sy.Asset(name="RSV Asset", description="Asset with RSV data", data_subjects=[rsv])
chuc_rsv_asset.set_obj(chuc_rsv_data)
chuc_rsv_asset.set_mock(chuc_rsv_mock, mock_is_real=True)
chuc_rsv_asset.set_shape(chuc_rsv_data.shape)
chuc_rsv_asset.add_data_subject(rsv)
chuc_dataset.add_asset(chuc_rsv_asset)
chuc_client.upload_dataset(chuc_dataset)
```

```python
chudsa_dataset = sy.Dataset(name="RSV Dataset", description="Dataset with RSV data")

chudsa_rsv_data = pd.read_csv("chuc.csv")
chudsa_rsv_mock = chudsa_rsv_data.sample(frac=0.1)

chudsa_rsv_asset = sy.Asset(name="RSV Asset", description="Asset with RSV data", data_subjects=[rsv])
chudsa_rsv_asset.set_obj(chudsa_rsv_data)
chudsa_rsv_asset.set_mock(chudsa_rsv_mock, mock_is_real=True)
chudsa_rsv_asset.set_shape(chudsa_rsv_data.shape)
chudsa_rsv_asset.add_data_subject(rsv)
chudsa_dataset.add_asset(chudsa_rsv_asset)
chudsa_client.upload_dataset(chudsa_dataset)
```
